```python
from google.cloud import bigquery
import pandas as pd
import plotly.express as px
```

```python
key_path = "../credentials/dbt_service_account.json"
client = bigquery.Client.from_service_account_json(key_path)
```

#### Pull visit counts by channel

```python
query = """
    SELECT traffic_medium, COUNT(*) AS visits
    FROM `amiable-dynamo-461319-g1.ga4_demo.int_sessions`
    GROUP BY traffic_medium
    ORDER BY visits DESC
"""
channel_df = client.query(query).to_dataframe()
```

#### Bar chart - channels by visits

```python

bar_fig = px.bar(
    channel_df,
    x="traffic_medium",
    y="visits",
    title=f"Traffic channels by visits",
    text="visits"
)

bar_fig.update_layout(
    xaxis_title="", 
    yaxis_title="visits",
    xaxis_tickangle=-45
)
bar_fig.show()
```

#### Pie chart (share of visits)

```python
pie_fig = px.pie(
    channel_df,
    names="traffic_medium",
    values="visits",
    title="Visit share by traffic channel",
    hole=0.3
)
pie_fig.update_traces(textposition="inside", textinfo="percent+label")
pie_fig.show()
```
